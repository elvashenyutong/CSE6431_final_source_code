```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df60 = pd.read_csv('Apple_60W_MagSafe_2_Power_Adapter.csv')
df85 = pd.read_csv('Apple_85W_MagSafe_2_Power_Adapter.csv')
dfsd = pd.read_csv('Apple_USB_SuperDrive.csv')
dfctosd = pd.read_csv('Apple_USB-C_to_SD_Card_Reader.csv')
df60['Date'] = pd.to_datetime(df60['Date'])
df85['Date'] = pd.to_datetime(df85['Date'])
dfsd['Date'] = pd.to_datetime(dfsd['Date'])
dfctosd['Date'] = pd.to_datetime(dfctosd['Date'])
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df60['Date'], df60['Sales'], marker='', linestyle='-', color='orange')

plt.title('Apple 60W MagSafe 2 PowerAdapter')
plt.xlabel('Date')
plt.ylabel('Sales')

plt.gcf().autofmt_xdate()

plt.savefig('Apple_60W_MagSafe_2_PowerAdapter.png')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df85['Date'], df85['Sales'], marker='', linestyle='-', color='orange')

plt.title('Apple 85W MagSafe 2 PowerAdapter')
plt.xlabel('Date')
plt.ylabel('Sales')

plt.gcf().autofmt_xdate()

plt.savefig('Apple_85W_MagSafe_2_PowerAdapter.png')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(dfsd['Date'], dfsd['Sales'], marker='', linestyle='-', color='orange')

plt.title('Apple USB SuperDrive')
plt.xlabel('Date')
plt.ylabel('Sales')

plt.gcf().autofmt_xdate()

plt.savefig('Apple_USB_SuperDrive.png')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(dfctosd['Date'], dfctosd['Sales'], marker='', linestyle='-', color='orange')

plt.title('Apple USB-C to SD Card Reader')
plt.xlabel('Date')
plt.ylabel('Sales')

plt.gcf().autofmt_xdate()

plt.savefig('Apple_USB-C_to_SD_Card_Reader.png')

plt.show()
```

```python
merged_df = pd.merge(df60, df85, on='Date', how='outer', suffixes=('_60', '_85'))

merged_df['Sales_60'].fillna(0, inplace=True)
merged_df['Sales_85'].fillna(0, inplace=True)

merged_df['Sales'] = merged_df['Sales_60'] + merged_df['Sales_85']

result_df = merged_df[['Date', 'Sales']]
result_df = result_df.sort_values(by='Date')

print(result_df.head())
```

```python
result_df
```

```python
plt.figure(figsize=(10, 6))
plt.plot(result_df['Date'], result_df['Sales'], marker='', linestyle='-', color='orange')

plt.title('Apple MagSafe 2 Power Adapter')
plt.xlabel('Date')
plt.ylabel('Sales')

plt.gcf().autofmt_xdate()

plt.savefig('Apple_MagSafe_2_Power_Adapter_comb.png')

plt.show()
```
